```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# .env 파일에서 환경 변수 로드
load_dotenv()

# OpenAI 클라이언트 생성 및 API 키 설정
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# 요약 생성 함수
def summarize_text_to_bullet(text, max_length=50):
    prompt = (
        f"다음 내용을 개조식으로 50자 이하의 핵심 문장으로 요약해 주세요:\n\n{text}\n\n"
        "요약된 결과:"
    )

    # 최신 OpenAI API 호출
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",  # 사용할 모델을 지정합니다
        messages=[
            {"role": "user", "content": prompt}
        ],
        max_tokens=max_length,  # 최대 토큰 수 설정
        temperature=0.3  # 생성의 다양성 조절
    )

    # 응답에서 요약 추출
    summary = response.choices[0].message['content'].strip()
    return summary

# 테스트 텍스트 요약
text = """
국내 최초의 등대공장, 물류 fulfillment 등의 이슈를 접하며, POSCO에 대한 관심을 가졌습니다. 
빠르게 디지털 전환을 성공시켜 smart-factory를 구축시킬 수 있었던 배경인 POSCO DX에 알게 되었습니다. 
계획, 제선, 제강, 연주, 압연, 도금으로 이루어진 생산방식과 물류관리 자동화 방식은 리딩 철강 브랜드로 만들기에 충분했다고 생각합니다. 
또한 글로벌 기업의 기술력이 아닌 사내 고유의 인프라를 활용해 신기술을 적용했기 때문에 확장성과 안정성을 동시에 갖출 수 있었습니다.
"""

# 요약 생성 및 출력
summary = summarize_text_to_bullet(text)
print("생성된 요약 (개조식, 50자 이하):", summary)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# 브라우저 드라이버 설정 (Chrome 예시)
driver = webdriver.Chrome()

# ChatGPT 웹사이트 열기 및 로그인
driver.get("https://chat.openai.com/")
time.sleep(5)  # 페이지 로딩 대기

# 추가적인 로그인 및 상호작용 자동화 (로그인 후에만 메시지 전송 가능)

# 메시지 입력 후 전송
message_box = driver.find_element("xpath", '//*[@placeholder="Send a message..."]')
message_box.send_keys("Python에서 ChatGPT와 상호작용하기")  # 메시지 입력
message_box.send_keys(Keys.RETURN)  # 메시지 전송
time.sleep(5)  # 응답 대기

# 응답 가져오기
response = driver.find_element("xpath", '//*[@class="message-content"]').text
print("ChatGPT 응답:", response)

# 브라우저 종료
driver.quit()
```
